```python
import pandas as pd

# Carregar os dados do arquivo Excel enviado pelo usuário
file_path = 'dados.xlsx'
data = pd.read_excel(file_path)

# Visualizar os dados carregados
data.head()
```

```python
# Definir a "Data PPH" como a data atual com a hora ajustada para 03:00
today_date = pd.Timestamp.today().normalize()  # remove a parte de tempo da data de hoje
data['Data PPH'] = today_date + pd.Timedelta(hours=3)

# Adicionar 3 horas à "Data PPH" para criar a coluna "Data ajustada" (mesma lógica, pois já está 03:00)
#data['Data ajustada'] = data['Data PPH']

# Aplicar a lógica para calcular a coluna "Saída"
data['Saída'] = data.apply(
    lambda row: 'Y' if pd.isna(row['Receiving.1']) or row['Receiving.1'] >= row['Data PPH'] else 'N',
    axis=1
)

# Mostrar o resultado
data[['Receiving.1', 'Data PPH', 'Saída']]
```

```python
import pandas as pd

# Carregar os dados do arquivo Excel enviado pelo usuário
file_path = 'dados.xlsx'
df = pd.read_excel(file_path)

# Visualizar os dados carregados
df.head()
```

```python
import pandas as pd
import numpy as np

def calculate_date(df):
    # Passo 1: Obter o valor de 'Data PPH' da primeira linha do DataFrame.
    date_value = pd.to_datetime(df.loc[0, 'Data PPH'])

    # Passo 2: Calcular o delta de tempo de 3 horas (0.125 dias)
    time_delta = pd.Timedelta(days=0.125)  # 0.125 dias equivale a 3 horas (0.125 * 24 horas)

    # Passo 3: Calcular a data ajustada
    adjusted_date = date_value + time_delta
    df['Data PPH'] = adjusted_date
    print(f"Data ajustada (adjusted_date): {adjusted_date}")

    # Passo 4: Garantir que 'Receiving.1' seja datetime64[ns]
    df['Receiving.1'] = pd.to_datetime(df['Receiving.1'], errors='coerce')

    # Passo 5: Filtrar as linhas que correspondem à data de hoje (sem parte do tempo)
    today = pd.Timestamp.today().normalize()  # Obtém a data atual sem a parte do tempo
    print(f"Data de hoje (today): {today}")

    # Passo 6: Adicionar a coluna 'Status' com a lógica de cálculo
    df['Status'] = np.where(
        (df['Receiving.1'].isna()) | (df['Receiving.1'] >= adjusted_date),  # Condição de comparação
        'Y',  # Se a condição for verdadeira, coloca 'Y'
        'N'   # Se a condição for falsa, coloca 'N'
    )

    return df

# Exemplo de uso
# Supondo que 'df' seja o seu DataFrame com as colunas 'Receiving.1' e 'Data PPH'


# Aplicando a função
df = calculate_date(df)

# Mostrar o DataFrame resultante
df[['Receiving.1', 'Data PPH', 'Status']]
```
